```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
data1 = pd.read_csv("no_interventions.csv", sep=",")
data1 = data1.drop(["Unnamed: 0"], axis=1)

data2 = pd.read_csv("A_(-2).csv", sep=",")
data2 = data2.drop(["Unnamed: 0"], axis=1)
```

```python
fig, axs = plt.subplots(4, 4)
fig.set_size_inches(35, 35)
fig.suptitle('Epic gamer moment')

columns = ['A', 'B', 'C', 'D']
for i in range(4):
    for j in range(4):
        axs[i,j].scatter(data1[columns[i]], data1[columns[j]])

plt.show()
```

```python
columns = ['A', 'B', 'C', 'D']

for i in range(4):
    sns.distplot(data1[columns[i]], kde=False, norm_hist=True, label='Without Intervention')
    sns.distplot(data2[columns[i]], kde=False, norm_hist=True, label='With Intervention')

    # Plot formatting
    plt.legend(prop={'size': 12})
    plt.title('Impact of Intervention')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```
