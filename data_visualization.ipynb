```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
data1 = pd.read_csv("data_real.csv", sep=",")
data1 = data1.drop(["Unnamed: 0"], axis=1)

data2 = pd.read_csv("data_489.csv", sep=",")
data2 = data2.drop(["Unnamed: 0"], axis=1)
```

```python
fig, axs = plt.subplots(6, 6)
fig.set_size_inches(10, 10)
fig.suptitle('100 samples: No intervention')

columns = ['A', 'B', 'C', 'D', 'E', 'F']
for i in range(6):
    for j in range(6):
        axs[i,j].scatter(data1[columns[i]], data1[columns[j]])

plt.show()
```

```python
fig, axs = plt.subplots(6, 6)
fig.set_size_inches(10, 10)
fig.suptitle('75 samples: A = 2')

columns = ['A', 'B', 'C', 'D', 'E', 'F']
for i in range(6):
    for j in range(6):
        axs[i,j].scatter(data2[columns[i]], data2[columns[j]])

plt.show()
```

```python
columns = ['A', 'B', 'C', 'D', 'E', 'F']

for i in range(6):
    sns.distplot(data1[columns[i]], kde=False, label='Without Intervention')
    sns.distplot(data2[columns[i]], kde=False, label='With Intervention')

    # Plot formatting
    plt.legend(prop={'size': 12})
    plt.title('Impact of Intervention')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```
