```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
datasets = ['A_-2', 'B_2', 'C_2', 'D_-2', 'E_-2', 'F_-2']
columns = ['A', 'B', 'C', 'D', 'E', 'F']
n = 6

fig, axs = plt.subplots(n, n)
fig.set_size_inches(35, 35)

for i in range(n):
    data = pd.read_csv("no_interventions.csv", sep=",")
    data = data.drop(["Unnamed: 0"], axis=1)
    data_intervened = pd.read_csv(datasets[i] + ".csv", sep=",")
    data_intervened = data_intervened.drop(["Unnamed: 0"], axis=1)
    for j in range(n):
        axs[i,j].hist(data[columns[j]], cumulative=True, label='Without Intervention', alpha=1, density=True, histtype="step", bins=100)
        axs[i,j].hist(data_intervened[columns[j]], cumulative=True, label='With Intervention', alpha=1, density=True, histtype="step", bins=100)
        
        # axs[i,j].set_xlabel('Value')
        # axs[i,j].set_ylabel('Density')

axs[0,0].legend(prop={'size': 12}, loc=2)

plt.show()
```
